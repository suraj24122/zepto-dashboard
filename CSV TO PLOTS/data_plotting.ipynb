```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv("12. the total inventory weight per category.csv")

# Plot
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='category', y='total_weight')
plt.xticks(rotation=45)
plt.title("Total Inventory Weight per Category")
plt.xlabel("Category")
plt.ylabel("Weight (grams)")
plt.tight_layout()
plt.show()
plt.savefig('PNG IMAGES/total inventory weight per category.png')
```

```python
df = pd.read_csv("5. top 10 best value products based on the discounted percentages.csv")

plt.figure(figsize=(10,6))
sns.barplot(data=df, y='name', x='discountPercent', palette='coolwarm')
plt.title("Top 10 Best Value Products (Highest Discounts)")
plt.xlabel("Discount (%)")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
plt.savefig('PNG IMAGES/Top 10 Best Value Products (Highest Discounts).png')
```

```python
df = pd.read_csv("7. estimated revenue for each category.csv")

plt.figure(figsize=(10,6))
sns.barplot(data=df, x='category', y='total_revenue')
plt.xticks(rotation=45)
plt.title("Estimated Revenue per Category")
plt.xlabel("Category")
plt.ylabel("Revenue (INR)")
plt.tight_layout()
plt.show()
plt.savefig('PNG IMAGES/Estimated Revenue per Category.png')
```

```python
df = pd.read_csv("8. products where mrp is greater than 500 and discount is less than 10%.csv")

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mrp', y='discountPercent')
plt.title("MRP vs Discounted Price (Low Discounts, High MRP)")
plt.xlabel("MRP")
plt.ylabel("Discounted Price")
plt.tight_layout()
plt.show()
plt.savefig('PNG IMAGES/MRP vs Discounted Price (Low Discounts, High MRP).png')
```

```python
df = pd.read_csv("9. top 5 categories offering the highest avg discount percentage.csv")

plt.figure(figsize=(8,5))
sns.barplot(data=df, x='category', y='avg_discount', palette='viridis')
plt.title("Top 5 Categories with Highest Avg Discount")
plt.ylabel("Average Discount (%)")
plt.tight_layout()
plt.show()
plt.savefig('PNG IMAGES/Top 5 Categories with Highest Avg Discount.png')
```
